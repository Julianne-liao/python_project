```python
#建立工廠
class Person:
    def __init__(self): #自訂的初始化
        print("初始化:一個新的人被建立了")
```

```python
p1 = Person()
type(p1)
```

```python
class Person:
    def __init__(self,name):
        #建立一個attribute屬性
        self.name = name

    #自訂實體的method
    def introduce(self):
        print(f"哈囉，我的名字是{self.name}")
```

```python
p2 = Person("robert")
p2.introduce()
```

```python
p3 = Person("jenny")
p3.introduce()
```
